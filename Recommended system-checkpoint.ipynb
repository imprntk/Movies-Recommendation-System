```python
import ast

def convert(text):
    l = []
    for i in ast.literal_eval(text):
        l.append(i['name'])
        
    return l
```

```python
def convert_cast(text):
    l = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            l.append(i['name'])
        counter += 1   
        
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
movies.head(2)
```

```python
import ast

def fectch_director(text):
    l = []
    for i in ast.literal_eval(text):
        if i["job"] == "Director":
            l.append(i['name'])
            break
        
        
    return l
```

```python
movies['crew'] = movies['crew'].apply(fectch_director)
```

```python
#sam worthington -> samworthington

def remove_space(word):
    l = []
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)


```

#vectorize 

```python
import nltk
from nltk.stem import PorterStemmer 
```

```python
ps = PorterStemmer()
```

```python
def stems(text):
    l = []
    for i in text.split():
        l.append(ps.stem(i))
        
    return " ".join(l)
```

```python
new_df['tags'] = new_df['tags'].apply(stems)
```

```python
new_df.iloc[0]['tags']
```

#counter vectorizing -> means removing meanigless words like is ,the,in etc.

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000,stop_words = 'english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

#getting the similairty movies

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)  #gives me similarity code
```

```python
similarity
```

```python
similarity.shape
```

#calculate my movie index 

```python
new_df[new_df['title'] == 'Spider-Man'].index[0]
```

#identify the index from dataset and checking distances or similar movies

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse = True, key = lambda x:x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Spider-Man')
```

```python
recommend('The Dark Knight Rises')
```
